```python
from gradio_client import Client, handle_file

# Ask the user for file paths
vton_img_path = input("Enter the path to the VTON image: ")
garm_img_path = input("Enter the path to the garment image: ")

# Create a client instance
client = Client("levihsu/OOTDiffusion")


result = client.predict(
    vton_img=handle_file(vton_img_path),
    garm_img=handle_file(garm_img_path),
    category="Lower-body",  # Specify the category as "Dress","Lower-body","Upper-body"
    n_samples=1,
    n_steps=20,
    image_scale=2,
    seed=-1,
    api_name="/process_dc"
)

# Print the result
print(result)
```
